```python
#importing the libraries
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Importing the data from the webpage into a DataFrame

url='https://en.wikipedia.org/wiki/List_of_highest-grossing_Russian_films'
req=requests.get(url)
data=pd.read_html(req.text)
df=data[0]
```

```python
df.head
```

```python
df
```

```python
print(df.to_string())
```

```python
df['Gross in Rubles']=df['Gross in Rubles'].str.replace(r"$","").str.replace(r",","")
```

```python
df.info()
```

```python
df['Gross in Rubles']=df['Gross in Rubles'].str.replace(r"$","").str.replace(r",","")
df['Gross in Rubles']=df['Gross in Rubles'].str.replace(r"\[[^\]]+\]","").str.replace(r",","")
```

```python
df['Gross in Rubles']=df['Gross in Rubles'].astype('int64')
```

```python
df.info()
```

```python
df["Gross in Rubles"].head()
```

```python
df
```

```python
df.drop(['Rank'],axis=1,inplace=True)
```

```python
df
```

```python
df1=df[['Title','Gross in Rubles']].sort_values(ascending=False,by='Gross in Rubles')
```

```python
df1
```

```python
#Plotting the top 5 films by revenue
#setting the figure size
plt.figure(figsize=(10,5))
#creating a bar plot
ax=sns.barplot(x='Title',y='Gross in Rubles',data=df1.head(5))
#rotating the x axis labels
ax.set_xticklabels(labels=df1.head()['Title'],rotation=75)
#setting the title
ax.set_title("Top 5 Films in Russia")
#setting the Y-axis labels
ax.set_ylabel("Gross in Rubles")
#Labelling the bars in the bar graph
for p in ax.patches:
  ax.annotate(p.get_height(),(p.get_x()+p.get_width()/2,p.get_height()),ha='center',va='bottom')
```

```python
df1['Gross in Rubles'].head(10).plot(kind='pie',autopct='%.2f%%',labels=df1['Title'],figsize=(10,5))
```

```python
df2=df.groupby('Year')['Gross in Rubles'].mean()
#creating a line plot
df2.plot(kind='line',figsize=(10,10))
```
